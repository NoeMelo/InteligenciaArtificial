## Obtener links a páginas 

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://es.wikipedia.org/wiki/Ciencias_de_la_computación'
```

```python
# Obtener el contenido
content = requests.get(url).content
```

```python
# Obtener soup
soup = BeautifulSoup(content,'lxml') # choose lxml parser
```

```python
# Encontrar todos los parrafos <p>
p_tags = soup.findAll('p')
```

```python
len(p_tags)
```

```python
# Obtener todos los tags <a>
a_tags = []
for p_tag in p_tags:
    a_tags.extend(p_tag.findAll('a'))
```

```python
a_tags
```

```python
len(a_tags)
```

```python
a_tags[0]
```

```python
# Filtrar solo referencias
a_tags = [ a_tag for a_tag in a_tags if 'title' in a_tag.attrs and 'href' in a_tag.attrs ]
```

```python
len(a_tags)
```

```python
a_tags
```

```python
for i in a_tags:
    print(i.get('href'))
```

```python
# Imprimir todos los links
for i,a_tag in enumerate(a_tags):
    print('[{0}] {1} -> {2}'.format(i,a_tag.get('title'),a_tag.get('href')))
```
