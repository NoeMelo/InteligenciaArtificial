## Obtener √çndice

```python
from bs4 import BeautifulSoup
import requests#conexion con la pagina web:   crawler python
```

```python
#URL inicial
#
url = 'https://es.wikipedia.org/wiki/Sevilla'
```

```python
# Obtener Links
content = requests.get(url).content
```

```python
content
```

```python
# Obtener soup
soup = BeautifulSoup(content,'lxml') # choose lxml parser, ESTRUCTURA LA P√ÅGINA EN FORMA DE ARBOL: HEAD -> link, title ,BODY:h1,h2,p,i,etc
```

```python
# Ubicar tags : <div class="toc">
tag = soup.find('div', {'class' : 'toc'}) # id="toc" also works
```

```python
tag
```

```python
# Obtener los links "<a>"
links = tag.findAll('a') # <a href='/path/to/div'>topic</a>
```

```python
links[2]
```

```python
len(links)
```

```python
links[0]
```

```python
links[0].text
```

```python
# Imprimir textos
for link in links:
  print(link.text) # Obtener los textos <a>
```
