Escriba una función <b>Func3</b> que tenga como entrada dos números enteros (INICIO,FIN) y que retorne todos los números que son divisibles por 7 pero que no son múltiplos de 5 entre INICIO y FIN (ambos incluídos). En la salida deben aparecer los números separados por comas. 

```python
import numpy as np
```

```python
def Func1(ini,fin):
    num = []
    for i in range(ini,fin):
        if i % 7 == 0 and i % 5 == 0:
            num.append(i)
    return num
```

[6 pts.] <b>Ejercicio 3:</b> 

En base al archivo <b>Email101.txt</b>, generar un archivo <b>Email102.txt</b> en el cual se realice el siguiente proceso de limpieza:

* Todas las palabras a minúsculas
* Mantener sólo palabras con más de 4 caracteres
* Eliminar dígitos [0-9]

```python
f = open("Email101.txt","r")
lines = f.readlines()
f.close()

f2 = open("Email102.txt","w")

for line in lines:
    selected =[]
    
    for word in line.split():
        if len(word)>4 and word.isalpha(): ## si es una palabra
            selected.append(word.lower())
    
    strg = " ".join(selected)
    
    f2.write(strg+"\n")
    
f2.close()
```

Usando el archivo <b>Email101.txt</b>, realizar la suma de todas las cantidades que están representado en dólares.

```python
#CODE
import re 

f = open("Email101.txt","r")
lines = f.readlines()
f.close()

strg = " ".join(lines)

lista = []
suma = 0
for w in strg.split():
    if w[0]=='$' and w[1:].isdigit():
        lista.append(w)
        suma = suma + int(w[1:])

print("Monto Total: "+str(suma))
print("Montos: ",lista)
```

```python
#CODE
I = np.random.randint(255, size=(10, 10))
F = np.random.randint(2, size=(3, 3))
```

```python
#CODE
n,m = I.shape

new_matrix=np.zeros((n,m))

for fila in range(1,n-1):
    
    for columna in range(1,m-1):
    
        matriz=I[fila-1:fila+2,columna-1:columna+2]*F
        
        new_matrix[fila,columna]=np.sum(matriz)
        
print(new_matrix)
```
